```python
import re
from collections import defaultdict, deque

def input():
   retval = defaultdict(list)
   compiled_pattern = re.compile(r'(\d+)\s(\d+)', re.ASCII)
   with open("/Users/evert/Downloads/SCC.txt", 'r', encoding='utf-8', newline='\n') as input_stream:
       for line in input_stream.readlines():
           match = compiled_pattern.match(line)
           if match:
               retval[int(match.group(2), base=10)].append(int(match.group(1), base=10))
   return retval

def reverse_of(V):
   retval = defaultdict(list)
   for src in V:
       for tgt in V[src]:
           retval[tgt].append(src)
   return retval

def dfs_1(V, starting_vertex, visited, timer):
   stack = deque([starting_vertex])
   while len(stack) > 0:
       src = stack[0]
       if src in visited:
           timer += 1
           visited[src] = timer
           stack.popleft()
       else:
           visited[src] = None
           for tgt in sorted(V[src]):
               if tgt not in visited:
                   stack.appendleft(tgt)
   return timer

def dfs_loop_1(V):
   visited = dict()
   timer = 0
   for starting_vertex in sorted(V.keys(), reverse=True):
       if starting_vertex not in visited:
           timer = dfs_1(V, starting_vertex, visited, timer)
   sort_by = lambda x: visited[x]
   return sorted(visited.keys(), key=sort_by, reverse=True)
   #dict([ [ordering[key], key] for key in visited])
   
def dfs_2(V, starting_vertex, visited):
   stack = deque([starting_vertex])
   while len(stack) > 0:
       src = stack.popleft()
       visited[src] = starting_vertex
       for tgt in sorted(V[src], reverse=True):
           if tgt not in visited:
               stack.appendleft(tgt)
   
   
def dfs_loop_2(V, order):
   visited = dict()
   for starting_vertex in order:
       if starting_vertex not in visited:
           dfs_2(V, starting_vertex, visited)
   retval = defaultdict(list)
   for v in visited:
       retval[visited[v]].append(v)
   return retval

print('Reading input...', end='')
V = input()
print(' done.')
#print('Reversing the graph...', end='')
#V = reverse_of(V)
#print(' done.')
print('Running DFS1...', end='')
new_order = dfs_loop_1(V)
print(' done.')
print('Reversing the graph...', end='')
V = reverse_of(V)
print(' done.')
print('Running DFS2...', end='')
result = dfs_loop_2(V, new_order)
print(' done.')
print(sorted([len(x) for x in result.values()], reverse=True))
```
